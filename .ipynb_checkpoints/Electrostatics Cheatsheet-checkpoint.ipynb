# Electrostatics Equation Cheatsheat  


## Coulomb's Force 
Original Coulomb's equation that relates the force between two charges and the distance between them.

### $F = k\frac{q_1q_2}{r^2}$

## Electric Field Formula  
This equation calculates the electric field produced by a charge onto another charge at a distance $r$.  
### $ E = \frac{F_e}{q_0} = k_e\frac{q_{src}}{r^2}$

## Electric Potential  
The electric potential difference is the change in potential energy per unit of charge in an electric field. $W$ is the work in Joules needed to move a charge $q$  Coulombs the noted distance. It is measured in Volts.  
### $\Delta V = \frac{W}{q}$  
It can also be written as:  
### $\Delta V = k\frac{q_{src}}{r}$

## Ohm's Law  

Ohm's law describes the relationship between voltage, current, and resistance.

#### $V= IR$  
This is the most common way it is found and solves for Voltage, but can be rearranged to suit a variety of situatiosn rather easily.
#### $I= \frac{V}{R}$  
This solves for current.
#### $R= \frac{V}{I}$
And this solves for resistance (load).

##### Voltage (V) is the electric potential, or the Potential Energy per unit charge.  
##### Current is how much charge flowing past a certain point, or charge over time $\frac{Q}{t}$.  
##### Resistance is how much the charge flow is impeded, denoted by the ohm $\Omega$.  

# Electrostatic Units  
#### 1 Volt = 1 Joule per Coulomb $\frac{J}{C}$

#### 1 Nm = 1 Joule

#### 1 Joule = 1 Nm  
(The energy it takes to move a 1 Newton object a distance of 1 meter)

#### 1 Coulomb = $6.25*10^{18}$ electrons or protons, resulting in each particle having about $1.6*10^{-19}$ coulombs each.

#### 1 Amp = 1 C per second at a given point - this is the unit used for current.

#### 1 Watt = 1 Joule per second $\frac{J}{s}$

#### 1 Ohm = 1 Volt per Amp $\frac{V}{A}$

## Electrical Circuit Components  


#### Resistors:  
Resistors, as the name implies, help restrict and control the voltages and currents in a circuit.  
#### Capacitors:  
Capcitors are often used to introduce a time-delay in a circuit. They're like batteries with very low capacity. Unlike many resistors, capacitors come both polarized and non-polarized. Poloarized capacitors must be placed in a circuit in a certain way, while non-polorized capcitors do not.
#### Transistor:  
In a nutshell, transistors are a sort of switch controlled by an electrical signal. They differ from actual switches by not just have an "On" or "Off" state, they can be modulated based on the current flowing through it.
#### Inductor:  
An inductor is just a coil of wire. Their purpose is to resist changes in current. In a simple circuit with just an LED and a resistor, introducing an inductor to it will prevent the LED from instantly switching on when power is supplied. Instead, the inductor resists the change in current and the LED comes on gradually until the maximum current is reached.
#### Integrated Circuits:  
Integrated circuits, or IC's are an electronic circuit shrunk down to fit inside of a chip. Different IC's have different purposes, and can have a wide variety in numbers of pins.
#### Cells:  
Cells are generally batteries or series of batteries that provide power to the circuit. The longer line of a cell represents the positive end.  
#### Switches:  
Switches are another broad category, but can be simplified for the purpose of this notebook to be split into toggles and buttons. Toggles are switches that close a circuit when flipped in a certain direction, while buttons close a circuit by being pressed. This is probably unnecessary information, but you never know when you might freeze up with information recall.
#### Ground or Earth:  
Electrical grounding is used as a way to prevent any electrical spikes in a circuit to be directed away and absorbed so as to not damage the circuit itself.
